```python
word_to_index_copy = word_to_index.copy()
for i in word_to_index.keys():
    if (len(i)==3):
        del(word_to_index_copy[i])
```

```python
count = Counter(word_to_index_copy)
```

```python
rand_count = random.choices(list(count.items()),k=100)
rand_count = dict(rand_count)
count2 = Counter(rand_count)
```

```python
tag = count2.most_common(80)
taglist = pytagcloud.make_tags(tag, maxsize=16)
df = pd.DataFrame(taglist)
df['size'] = ((df['size']-min(df['size'])+5)**(1.5)).astype(int)
taglist2 = df.to_dict('records')
taglist2
```

```python
pytagcloud.create_tag_image(taglist2, 'result/고객상담.jpg', size=(800, 600), fontname='korean', rectangular=False)
```
